```python
import torch
import torch.nn as nn
import torch.optim as optim
import numpy as np 

%matplotlib inline
%pylab inline

print(torch.__version__)
```

```python
class Line2D(nn.Module):
    def __init__(self):
        super(Line2D, self).__init__()
        self.a = nn.Parameter(torch.rand(1))
        self.b = nn.Parameter(torch.rand(1))

    def forward(self, x):
        # x: 1xn (n data points, and the data point dimension is 1)
        y = self.a * x + self.b
        return y 
```

```python
# generate data 
data_x = torch.linspace(0, 100, 300)

noise_max = 1
true_a = 1.8
true_b = -30
data_y = (true_a*data_x + true_b) \
       + noise_max*torch.rand(data_x.shape)

# add outlier noise 
ratio_outliers = 0.9
num_outliers = round(ratio_outliers*data_x.shape[0])
outliers_idx = np.random.choice(data_x.shape[0], num_outliers)

jitter = torch.rand(round(num_outliers)) + (2*torch.rand(round(num_outliers))- 1)
outlier_noise_max = 300
data_y[outliers_idx] = data_y[outliers_idx] \
                     + outlier_noise_max*jitter # to make the jitter btn [-1.25, .75]
```

```python
# the initial predicted line 
line2d = Line2D() # init model 
pred_y = line2d(data_x).detach().numpy()

# draw the initial prediction before the optimization 
fig = plt.figure(1)
ax = fig.add_subplot(111)
plt.scatter(data_x, data_y, color='black', s=10, label='truth (y=0.5x)')
plt.plot(data_x, pred_y, color='r', linewidth=2, label='pred (initial)')
title('toy data with noise')
xlabel('x-axis')
ylabel('y-axis')
legend(loc='lower right')
# ax.set_aspect('equal', adjustable='box')
```

```python
"""
    loss 1: a gnereal non-robust loss  
"""
# re-init model 
line2d = Line2D() 
print('current parameters - a:', line2d.a.data.numpy(), 'b:', line2d.b.data.numpy())

# optimizer 
optimizer = optim.Adam(line2d.parameters(), lr=0.01)
scheduler = optim.lr_scheduler.StepLR(optimizer, step_size=10000, gamma=0.8)

# loss 
def L1loss(pred, true):
    return torch.abs(true - pred).mean()

# run optimization 
l1loss_history = []
for i in range(30000):
    optimizer.zero_grad()
    pred_y = line2d(data_x)
    loss = L1loss(pred_y, data_y)
    loss.backward()
    optimizer.step()
    scheduler.step()

    l1loss_history.append(loss.detach().numpy())
    if i % 5000 == 0:
        loss, current = loss.item(), i * len(data_x)
        print(f"loss: {loss:>10f}")
        

"""
    draw the result 
"""
print('optimized parameters - a:', line2d.a.data.numpy(), 'b:', line2d.b.data.numpy())
print('true parameters      - a:', true_a, 'b:', true_b)

# draw the optimized prediction after the optimization 
pred_y_after_opt = line2d(data_x).detach().numpy()

fig = plt.figure(2)
ax = fig.add_subplot(111)
plt.scatter(data_x, data_y, color='black', s=10, label='truth (y=0.5x)')
plt.plot(data_x, pred_y_after_opt, color='r', linewidth=2, label='pred (optimized)')
title('non-robust optimization result')
xlabel('x-axis')
ylabel('y-axis')
legend(loc='lower right')
# ax.set_aspect('equal', adjustable='box')

fig = plt.figure(3)
ax = fig.add_subplot(111)
plt.plot(l1loss_history)
xlabel('# iterations')
ylabel('L1 loss')
```

```python
"""
    loss 2: the robust huber loss   
"""
# re-init model 
line2d = Line2D() 
print('current parameters - a:', line2d.a.data.numpy(), 'b:', line2d.b.data.numpy())

# optimizer 
optimizer = optim.Adam(line2d.parameters(), lr=0.01)
scheduler = optim.lr_scheduler.StepLR(optimizer, step_size=10000, gamma=0.8)

# run optimization 
huberloss_history = []
for i in range(30000):
    optimizer.zero_grad()
    pred_y = line2d(data_x)
    loss = nn.HuberLoss(delta=0.1)(pred_y, data_y)
    loss.backward()
    optimizer.step()
    scheduler.step()
    
    huberloss_history.append(loss.detach().numpy())
    if i % 5000 == 0:
        loss, current = loss.item(), i * len(data_x)
        print(f"loss: {loss:>10f}")
        

"""
    draw the result 
"""
print('optimized parameters - a:', line2d.a.data.numpy(), 'b:', line2d.b.data.numpy())
print('true parameters      - a:', true_a, 'b:', true_b)

# draw the optimized prediction after the optimization 
pred_y_after_opt = line2d(data_x).detach().numpy()

fig = plt.figure(2)
ax = fig.add_subplot(111)
plt.scatter(data_x, data_y, color='black', s=10, label='truth (y=0.5x)')
plt.plot(data_x, pred_y_after_opt, color='r', linewidth=2, label='pred (optimized)')
title('non-robust optimization result')
xlabel('x-axis')
ylabel('y-axis')
legend(loc='lower right')
# ax.set_aspect('equal', adjustable='box')

fig = plt.figure(3)
ax = fig.add_subplot(111)
plt.plot(huberloss_history)
xlabel('# iterations')
ylabel('Huber loss')
```
