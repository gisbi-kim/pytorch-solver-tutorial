```python
import torch
import torch.nn as nn
import torch.optim as optim
import numpy as np 

%matplotlib inline
%pylab inline

print(torch.__version__)
```

```python
# generate data 
data_x = torch.linspace(0, 10, 100)

noise_max = 1
data_y = 0.5*data_x + noise_max*torch.rand(data_x.shape)
```

```python
class Line2D(nn.Module):
    def __init__(self):
        super(Line2D, self).__init__()
        self.a = nn.Parameter(torch.rand(1))
        self.b = nn.Parameter(torch.rand(1))

    def forward(self, x):
        # x: 1xn (n data points, and the data point dimension is 1)
        return self.a * x + self.b 
```

```python
# init model 
line2d = Line2D()
```

```python
# initial predicted line 
pred_y = line2d(data_x).detach().numpy()

# draw the initial prediction before the optimization 
fig = plt.figure(1)
ax = fig.add_subplot(111)
plt.plot(data_x, data_y, color='black', linewidth=2, label='truth (y=0.5x)')
plt.plot(data_x, pred_y, color='r', linewidth=2, label='pred (initial)')
title('toy data with noise')
xlabel('x-axis')
ylabel('y-axis')
legend()
ax.set_aspect('equal', adjustable='box')
```

```python
# optimizer 
optimizer = optim.Adam(line2d.parameters(), lr=0.0001)
```

```python
# run optimization 
for i in range(30000):
    optimizer.zero_grad()
    pred_y = line2d(data_x)
    loss = torch.abs(data_y - pred_y).mean()
    loss.backward()
    optimizer.step()

    if i % 5000 == 0:
        loss, current = loss.item(), i * len(data_x)
        print(f"loss: {loss:>10f}")
```

```python
# draw the optimized prediction after the optimization 
pred_y_after_opt = line2d(data_x).detach().numpy()

fig = plt.figure(2)
ax = fig.add_subplot(111)
plt.plot(data_x, data_y, color='black', linewidth=2, label='truth (y=0.5x)')
plt.plot(data_x, pred_y_after_opt, color='r', linewidth=2, label='pred (optimized)')
title('toy data with noise')
xlabel('x-axis')
ylabel('y-axis')
legend()
ax.set_aspect('equal', adjustable='box')
```
